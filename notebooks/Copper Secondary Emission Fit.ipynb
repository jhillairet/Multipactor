# The semi-empirical law derivation
Let $n(z,E_{PE})$ be the rate at which Secondary Electrons (SE) are produced by one Primary Electron (PE) at a depth $z$ below the surface and  $f(z)$ the probability of escape for a secondary electron. The SE rate $\delta\left(E_{PE}\right)$ is the integral over the full thickness of the sample:
$$ 
\delta\left(E_{PE}\right) = \int_0^\infty n(z,E_{PE}) f(z) dz 
$$
As the incident electrons have a small penetration depth, only a small portion of the integrand is non-zero. 

We assume a constant stopping power, i.e.:
$$ n(z,E_{PE})= - \frac{1}{B} \frac{dE}{dz} = - \frac{1}{B} \frac{E_{PE}}{R} $$
where $B$ is the effective energy required to produce a SE and $R$ is the penetration depth of the incident primary electron. 


The probability function $f(z)$ that a generated SE will escape back to incident surface is assumed to be of the form:
$$ f(z)= K \exp\left( - \frac{z}{\lambda} \right) $$
where $\lambda$ is the effective SE escape depth (also defined as $1/\alpha$) and $K$ a constant.

This leads to:
$$
\delta\left(E_{PE}\right) = K \frac{E_{PE}}{B} \frac{\lambda}{R} \left(1  - e^{ - \frac{R}{\lambda} } \right)
$$

# curve fit

```python
def delta(E, B, eta, A, lambda0, n):
    # Rapport Nicolas
    # Marche pas..?
    R=E**(n+1)/((n+1)*A)
    delta = B/eta*(A*n*lambda0)**(1/n)*(R/lambda0)**(1/(n-1))*(1-np.exp(-R/lambda0))
    
    # J.J.Scholtz et al, 1996, Secondary Electron Emission Properties
    R=E**n/(A*n)
    delta=B/eta*E/R*lambda0*(1-np.exp(-R/lambda0))
    
    # Y.Lin and D.C.Joy
    
    
    return delta
    
```

```python
from scipy.optimize import curve_fit
```

```python
x0 = (0.79, 33.5, 30.5, 10.5, 1.25)
#x0 = (0.2, 15.5, 15.5, 10.5, 1.25)

try:
    popt, pcov = curve_fit(delta, E*1e3, SEY, x0)
    print(popt)
except OptimizeWarning:
    print('be')
except ValueError:
    print('puf')
```

```python
plot(E, SEY, lw=2)
plot(E, delta(E*1e3, popt[0], popt[1], popt[2], popt[3], popt[4]), lw=2)
grid(True)
legend(('[Joy] dataset #14', 'best fit'), fontsize=14)
```

# leastsq

```python
from scipy.optimize import leastsq
```

```python
def delta2(x, E_data, SEY_data):
    return(SEY_data - delta(E_data, x[0], x[1], x[2], x[3], x[4]))

x0 = (0.79, 33.5, 30.5, 10.5, 1.25)
x = leastsq(delta2, x0, args=(E*1e3, SEY))
print(x[0])
```

```python
plot(E, SEY)
plot(E, delta(E*1e3, x[0][0], x[0][1], x[0][2], x[0][3], x[0][4]))
```
